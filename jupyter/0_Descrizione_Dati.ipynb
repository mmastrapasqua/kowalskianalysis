```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.constants import golden

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize=(5.0, 5.0/golden))
```

```python
tragitti = pd.read_csv('csv/2020_from_03_01_to_03_14_moovit.csv', index_col=False)
```

```python
# Numero totale di tragitti
len(tragitti)
```

```python
tragitti.head()
```

```python
# Massima e minimia distanza di un tragitto
min = int(tragitti['Length'].min())
max = int(tragitti['Length'].max())
print(min, max) # km
```

```python
# Distribuzione in base alla lunghezza delle tratte
tragitti['Length'].hist(bins=np.hstack((np.arange(min, max+1, 1))))
plt.show()
```

```python
tragitti['Length'].plot.box(vert=False, whis='range')
plt.show()
```

```python
# Correlazione tra tragitto in macchina propria (macchina) e car sharing (piedi + macchina)
tragitti.plot.scatter('Car', 'Car2Go')
plt.show()
```

```python
tragitti.plot.scatter('Car', 'Enjoy')
plt.show()
```

```python
# Correlazione tra lunghezza tragitto e tempo di percorrenza coi vari mezzi
tragitti[tragitti['Car'] != 0].plot.scatter('Length', 'Car')
plt.show()
```

```python
tragitti[tragitti['Foot'] != 0].plot.scatter('Length', 'Foot')
plt.show()
```

```python
tragitti[tragitti['Bike'] != 0].plot.scatter('Length', 'Bike')
plt.show()
```

```python
tragitti[tragitti['Public'].between(1,100)].plot.scatter('Length', 'Public')
plt.show()
```

```python
tragitti[tragitti['Public2'].between(1,100)].plot.scatter('Length', 'Public2')
plt.show()
```
